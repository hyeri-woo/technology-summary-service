```python
import os
import glob
import json

current_path = os.getcwd()

# 현재 경로에서 data/training 폴더로 이동
folder_path = os.path.join(current_path, 'data/training')

# 해당 폴더 내의 모든 json 파일 경로 가져오기
json_files = glob.glob(os.path.join(folder_path, "*.json"))
json_files
```

```python
# context와 summary를 저장할 리스트
extracted_data = []

# JSON 파일 가공
for file in json_files:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
        # 'dataset' 키 아래의 'context_info' 접근
        context_info = data.get('dataset', {}).get('context_info', [])
        
        # context_info에서 각 아이템의 context와 summary 추출
        for item in context_info:
            context = item.get('context', "")
            summary = item.get('summary', "")
            extracted_data.append([{context, summary}])

print(extracted_data)
```
